```python
from  bs4 import BeautifulSoup #pip install beautifulsoup4
import requests
```

```python
url = "https://www.lushusa.com/face/cleansers-scrubs/?cgid=cleansers-scrubs&start=0&sz=28"
```

```python
# creating a session
r_session = requests.session()
r = r_session.get(url=url)
print(r)
```

```python
# defining a soup
soup = BeautifulSoup(r.text, 'html.parser')
soup
```

```python
# css for product_name
product_name_css = 'div.d-flex div.product-tile h3.product-tile-name'
product_names = soup.select(product_name_css)
# print(product_names)
product_names[1].text.strip()
```

```python
# getting attribute from html using 'get('<attrib_name>')'

product_names[0].select('a')[0].get('href')
```

```python
# getting description
product_desc_css = 'div.d-flex div.product-tile-tagline'
product_descs = soup.select(product_desc_css)
product_descs[0].text
```

```python
product_price_css = 'div.d-flex div.tile-price-size span.tile-price'
product_prices_list = soup.select(product_price_css)
product_prices_list
```
