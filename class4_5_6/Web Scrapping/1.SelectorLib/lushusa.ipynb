# https://www.lushusa.com

```python
from selectorlib import Extractor #pip install selectorlib
import requests
import pandas as pd 
r = requests.get('https://www.lushusa.com/face/cleansers-scrubs/?cgid=cleansers-scrubs&start=0&sz=28')
yml_text = """
product_data:
    css: 'div.d-flex div.product-tile'
    xpath: null
    multiple: true
    type: Text
    children:
        image:
            css: 'img.product-tile-image'
            xpath: null
            multiple: false
            type: Image
        name:
            css: 'h3.product-tile-name'
            xpath: null
            multiple: false
            type: Text
        price:
            css: 'div.tile-price-size span.tile-price'
            xpath: null
            multiple: false
            type: Text"""
e = Extractor.from_yaml_string(yml_text)
data = e.extract(r.text)
# print(data['product_data'])
df = pd.DataFrame(data['product_data'])
df
```

# https://bagallery.com

```python
import requests
from selectorlib import Extractor
all_data = []
page = 1
while True:
    url = f'https://bagallery.com/search?filter=1&page={page}&q=sunsilk&type=product'
    session = requests.Session()
    response = session.get(url = url)
    yaml = '''
    Product_data:
        css: 'div.grid-item div.product-bottom'
        xpath: null
        multiple: true
        type: Text
        children:
            name:
                css: 'a.product-title'
                xpath: null
                multiple: true
                type: Text
            actual_price:
                css: 'span.money'
                xpath: null
                multiple: true
                type: Text

    '''
    e = Extractor.from_yaml_string(yaml)
    data = e.extract(response.text)
    
    print(data['Product_data'])
    page += 1
    try:
        all_data += data['Product_data']
    except:
        print(page)
        break
```
