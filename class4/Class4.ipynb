```python
# loading xml file
tree = ET.parse('xmltest.xml')

# getting root from tree
root = tree.getroot()

for child in root.iter():
    print(child.tag)
```

```python
# all the elements in the entire tree
print([elem.tag for elem in root.iter()])
```

```python
for child in root.findall('genre')[0].find('decade'):
     print(child.tag, child.attrib)
```

```python
ET.tostring(root)
print(ET.tostring(root).decode('utf8'))
```

```python
# iterate over all the elements with `movie` as a object that have different properties
# .text if element contains any value
for a in root.iter():
    print("TAG: ",a.tag, "ATTR: ",a.attrib,"TEXT: " ,a.text)
```

```python
# list specific element attributes.
for movie in root.iter('genre'):
    print(movie.attrib)
```

```python
# XPath Expressions
'''
Printing out the XML is helpful, but XPath is a query language 
used to search through an XML quickly and easily.'''

for description in root.iter('description'):
    print(description.text)
```

```python
# condition based on value in node element
for movie in root.findall("./genre/decade/movie/[year='1984']"):
    print(movie.attrib)
```

```python
'''
to move up the parents use `..` to move one element above
the selected element `....` to above 2 elements
'''
for movie in root.findall("./genre/decade/movie/format[@multiple='Yes']...."):
    print(movie.tag, movie.attrib)
```
